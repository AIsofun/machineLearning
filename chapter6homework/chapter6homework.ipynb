```
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

train = pd.read_csv('./train_final.csv', engine='python')
test = pd.read_csv('./test_final.csv', engine='python')

train_x = np.array(train.drop(['loan_status'],1))
#print(train_x)
train_y = train.iloc[:,15]
#print(train_y)
test_x = np.array(test.drop(['loan_status'],1))
#print(test_x)
test_y = test.iloc[:,15]
#print(test_y)


```

```
import lightgbm as lgb 
lgb_train = lgb.Dataset(train_x, train_y)  
lgb_test = lgb.Dataset(test_x, test_y, reference=lgb_train)  

for val in [2, 4, 8, 16, 32, 64, 128]:
  params = {'boosting_type': 'gbdt',
        'num_thread': 4, 
        'num_leaves': 16, 
        'metric': 'binary', 
        'objective': 'binary',
        'num_round': 4000, 
        'learning_rate': 0.02,
        'feature_fraction': 0.8, 
        'bagging_fraction': 0.8,
        'max_depth': val}
  ret = lgb.train(params,lgb_train,num_boost_round=10000,valid_sets=lgb_test,early_stopping_rounds=500) 
  from sklearn import metrics
  from sklearn.metrics import accuracy_score, roc_auc_score
  pre = ret.predict(test.drop(columns='loan_status'),num_iteration = ret.best_iteration)
  for i in range(0, np.size(pre)):
      if pre[i] >= 0.5:
        pre[i] = 1
      else:  
        pre[i] = 0
  accuracy = accuracy_score(test_y, pre)
  print(params) 
  print('准确度：',accuracy) 
```
